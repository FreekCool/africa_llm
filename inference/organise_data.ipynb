```python
import os
import pandas as pd
```

```python
raw_data_dir = "raw_data"

csv_files = [f for f in os.listdir(raw_data_dir) if f.endswith(".csv")]
dfs = {f: pd.read_csv(os.path.join(raw_data_dir, f)) for f in csv_files}

for name, df in dfs.items():
    print(f"{name}: {df.shape}")
    display(df.head())
```

```python
for name in dfs:
    before = len(dfs[name])
    dfs[name] = dfs[name].dropna(subset=["text"])
    after = len(dfs[name])
    print(f"{name}: dropped {before - after} NaN rows, {after} remaining")
```

```python
df = pd.concat(dfs.values(), ignore_index=True).sample(frac=1, random_state=42).reset_index(drop=True)
print(f"Merged & shuffled: {df.shape}")
display(df.head())
```

```python
os.makedirs("data", exist_ok=True)
df.to_csv("data/inference_data.csv", index=False)
print("Saved to data/inference_data.csv")
```
