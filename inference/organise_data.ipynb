```python
import os
import pandas as pd
```

```python
raw_data_dir = "raw_data"

csv_files = [f for f in os.listdir(raw_data_dir) if f.endswith(".csv")]
dfs = {f: pd.read_csv(os.path.join(raw_data_dir, f)) for f in csv_files}

for name, df in dfs.items():
    print(f"{name}: {df.shape}")
    display(df.head())
```
