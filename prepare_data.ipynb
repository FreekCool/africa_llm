```python
import pandas as pd
import json
import os
```

```python
data_path = '/projects/prjs1308/africa_llm_data/for-freek-africa-fb-posts-to-annotate-1oct2025-KEY-VARS.xlsx'
df = pd.read_excel(data_path)
```

```python
# --- map: JSON key -> dataframe column name
# Keep the JSON keys aligned with the prompt structure.
col_map = {
    "id": "id",
    "post_owner.id": "post_owner.id",
    "text": "text",
    # new: language target
    "language": "language",
    "politics": "politics",
    "domestic_politics": "domestic_politics",
    "foreign_politics": "foreign_politics",
    "resource_distribution": "resource_distribution",
    "resource_distribution_by_whom1": "resource_distribution_by_whom1",
    "resource_distribution_for_whom_ethnic1": "resource_distribution_for_whom_ethnic1",
    "resource_distribution_for_whom_region1": "resource_distribution_for_whom_region1",
    "resource_distribution_for_gender": "resource_distribution_gender",
    "climate_change": "climate_change",
    "topic01": "topic01",
    "pro_us": "pro_us",
    "pro_russia": "pro_russia",
    "pro_china": "pro_china",
    "pro_un": "pro_un",
    "pro_mf": "pro_imf",
    "pro_democracy": "pro_democracy",
    "anti_western": "anti_western",
    "national_unity": "national_unity",
    "subgroup_unity": "subgroup_unity",
    "subgroup_unity_text": "subgroup_unity_text",
    "african_unity": "african_unity",
    # new: additional binary targets in the prompt
    "political_opponents": "political_opponents",
    "religion": "religion",
}

# --- select existing columns only (so it won't crash if something is missing)
existing = {k: v for k, v in col_map.items() if v in df.columns}
missing = [k for k, v in col_map.items() if v not in df.columns]

df_out = df[list(existing.values())].rename(columns={v: k for k, v in existing.items()})
records = df_out.to_dict(orient="records")

if missing:
    print("Missing (not found in df.columns):", missing)

# --- output folder
out_dir = "/projects/prjs1308/africa_llm_data/africa_jsons"
os.makedirs(out_dir, exist_ok=True)

# --- write files
jsonl_path = os.path.join(out_dir, "african_videos.jsonl")
json_path  = os.path.join(out_dir, "african_videos.json")

with open(jsonl_path, "w", encoding="utf-8") as f:
    for rec in records:
        f.write(json.dumps(rec, ensure_ascii=False) + "\n")

with open(json_path, "w", encoding="utf-8") as f:
    json.dump(records, f, ensure_ascii=False, indent=2)

print("Saved:", jsonl_path)
print("Saved:", json_path)
```
